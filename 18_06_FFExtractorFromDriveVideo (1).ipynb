```python
from scipy.spatial import distance as dist


#Code to calculate eye_aspect_ratio

def eye_aspect_ratio(eye):
	# compute the euclidean distances between the two sets of
	# vertical eye landmarks (x, y)-coordinates
	A = dist.euclidean(eye[1], eye[5])
	B = dist.euclidean(eye[2], eye[4])

	# compute the euclidean distance between the horizontal
	# eye landmark (x, y)-coordinates
	C = dist.euclidean(eye[0], eye[3])

	# compute the eye aspect ratio
	ear = (A + B) / (2.0 * C)

	# return the eye aspect ratio
	return ear
```

```python
def mouth_aspect_ratio(mouth):
	# compute the euclidean distances between the two sets of
	# vertical mouth landmarks (x, y)-coordinates
	A = dist.euclidean(mouth[2], mouth[10]) # 51, 59
	B = dist.euclidean(mouth[4], mouth[8]) # 53, 57

	# compute the euclidean distance between the horizontal
	# mouth landmark (x, y)-coordinates
	C = dist.euclidean(mouth[0], mouth[6]) # 49, 55

	# compute the mouth aspect ratio
	mar = (A + B) / (2.0 * C)

	# return the mouth aspect ratio
	return mar
```

```python
import cv2
import dlib
import pandas as pd
import imutils
from imutils.video import VideoStream
from imutils import face_utils,resize

# initialize dlib's face detector (HOG-based) and then create
# the facial landmark predictor
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("/content/drive/My Drive/ProyectoSaturdaysAI - IV Edición/shape_predictor_68_face_landmarks.dat")

# grab the indexes of the facial landmarks for the left and
# right eye, respectively
(lStart, lEnd) = face_utils.FACIAL_LANDMARKS_IDXS["left_eye"]
(rStart, rEnd) = face_utils.FACIAL_LANDMARKS_IDXS["right_eye"]
(mStart, mEnd) = face_utils.FACIAL_LANDMARKS_IDXS["mouth"]

for filePath in array:
  cap = cv2.VideoCapture(filePath)
  totalframecount= int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
  print(filePath)
  print("Frames per second ",cap.get(cv2.CAP_PROP_FPS))
  print("The total number of frames in this video is ", totalframecount)
  Y_status=os.path.splitext(filePath)[0].split("/")[-1]
  people_number=os.path.dirname(filePath).split("/")[-1]
  dataframe = pd.DataFrame()

  cols = [str(n) for n in range(0, 68)]
  df_points = pd.DataFrame(columns=cols)
  FRAME_COUNTER = 0
  FACES_NOT_FOUND = 0
  SKIP = 3
  FRAMES_PER_VIDEO_TRAIN=10000
  while FRAME_COUNTER <= FRAMES_PER_VIDEO_TRAIN:
    
    FRAME_COUNTER += 1
    if FRAME_COUNTER % int(totalframecount/2) == 0:
      print(str((FRAME_COUNTER/totalframecount)*100)+"% done")
    if (FRAME_COUNTER % SKIP) ==0:
      ret, frame = cap.read()
      if not ret:
        break
      # Convert image into grayscale
      frame = imutils.resize(frame, width=450)
      gray = cv2.cvtColor(src=frame, code=cv2.COLOR_BGR2GRAY)
      gray=check_rotation(str(people_number)+"_"+str(Y_status),gray)
      faces = detector(gray)
      if len(faces)==0:
        FACES_NOT_FOUND +=1
      if FACES_NOT_FOUND>=500:
        cv2.imwrite('/content/drive/My Drive/SaturdaysAI/dataSet/no_faces/'+str(people_number)+"_"+str(Y_status)+'.jpg', gray)  
        break
      for face in faces:
        # Create landmark object
        landmarks = predictor(image=gray, box=face)
        shape = face_utils.shape_to_np(landmarks)
        
        leftEye = shape[lStart:lEnd]
        rightEye = shape[rStart:rEnd]
        leftEAR = eye_aspect_ratio(leftEye)
        rightEAR = eye_aspect_ratio(rightEye)
        # average the eye aspect ratio together for both eyes
        ear = (leftEAR + rightEAR) / 2.0
        mar= mouth_aspect_ratio(shape[mStart:mEnd])
        face_featu_dict = {}
        # Loop through all the points
        face_featu_dict['EAR']=ear
        face_featu_dict['MAR']=mar
        face_featu_dict['FRAME']=FRAME_COUNTER
        face_featu_dict['Y']=Y_status
        
        for n in range(0, 68):
          x = landmarks.part(n).x
          y = landmarks.part(n).y
          # Saving the coordinates from all the points in a dictionary
          face_featu_dict[str(n)+'_x'] = x
          face_featu_dict[str(n)+'_y'] = y
        #adding a row with the info about all the points
        dataframe=dataframe.append(face_featu_dict, ignore_index=True)
  
  # saving the df with all the info about the points
  print("Faces Not Found:"+str(FACES_NOT_FOUND))
  
  if not dataframe.empty:
    print("Escribo csv")
    dataframe.to_csv('/content/drive/My Drive/SaturdaysAI/dataset/'+str(people_number)+"_"+str(Y_status)+".csv")
  # When everything done, release the video capture and video write objects
  cap.release()
  print("----------")
```

```python
from numpy import dstack
import numpy as np
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from math import sqrt

def scale_data_min_max(data):
  scaler = MinMaxScaler(feature_range=(0, 1))
  scaler = scaler.fit(data)
  data_norm = scaler.transform(data)
  return data_norm

def scale_data_standard(data):
  scaler = StandardScaler()
  scaler = scaler.fit(data)
  data_standarized = scaler.transform(data)
  return data_standarized

def load_group(filenames,dirs,Y):
  list_ear = list()
  list_mar = list()
  for filename in filenames:
    print(filename)
    if filename is not None and filename.lower().endswith('csv'):
      data = load_file('/content/drive/My Drive/SaturdaysAI/dataset/EAR/'+filename)
      data_mar = load_file('/content/drive/My Drive/SaturdaysAI/dataset/MAR/'+filename)
      data=scale_data_min_max(data)
      data_mar=scale_data_min_max(data_mar)
      data=scale_data_standard(data)
      data_mar=scale_data_standard(data_mar)
      Y.append(filename.split('.')[0].split('_')[1])
      list_ear.append(data)
      list_mar.append(data_mar)
  return list_ear,list_mar



def load_file(filepath):
  dataframe = pd.read_csv(filepath, header=None, delim_whitespace=True)
  #print(dataframe.shape)
  return dataframe.values


#Leemos ficheros y vamos almacenando al final tendremos 3 dimensiones rows,128,2
filenames=[]
get_videos_filepath('/content/drive/My Drive/SaturdaysAI/dataset/EAR',filenames)
dirs=['EAR','MAR']
Y=[]

list_ear,list_mar=load_group(filenames,dirs,Y)

print(len(Y))
print(len(list_ear))
print(len(list_mar))

print(len(list_ear[1]))
print(len(list_mar[1]))

y_final=[]
features_ear=[]
features_mar=[]
counter=0
for obj in list_ear:
  for sub_obj in obj:
    features_ear.append(sub_obj)
    features_mar.append(sub_obj)
    y_final.append(Y[counter])
  counter +=1
print(len(features_ear))
print(len(features_mar))
print(len(y_final))
features_ear=scale_data_min_max(features_ear)
features_ear=scale_data_standard(features_ear)
features_mar=scale_data_min_max(features_mar)
features_mar=scale_data_standard(features_mar)
print(features_ear[0].shape)
print(features_mar.shape)
print('---------')
matrix=np.zeros([len(y_final),32,2])
for x in range(0,len(y_final)):
  #print(features[x])
  #vector_ear=np.expand_dims(features_ear, axis=1)
  #vector_mar=np.expand_dims(features_mar, axis=1)
  vector_combined=np.vstack((features_ear[x], features_mar[x])).T
  matrix[x]=vector_combined
print(matrix.shape)
print(y_final[0])
print(y_final[3])
y_final=[0 if x == '0' else x for x in y_final]
y_final=[1 if x == '5' else x for x in y_final]
y_final=[2 if x == '10' else x for x in y_final]

y_final=to_categorical(y_final)
print(y_final[0])
print(y_final[3])

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(matrix, y_final, test_size=0.20, random_state=46)
```

```python
from keras.models import Sequential
from keras.layers import Bidirectional,Activation
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import Dropout
from keras.layers import LSTM
from keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras.optimizers import SGD

verbose, epochs, batch_size = 1, 2500, 32
n_timesteps, n_features, n_outputs = 64, 2, 3
model = Sequential()
model.add(LSTM(100, return_sequences=True, input_shape=(32,2),dropout=0.4))
model.add(Flatten())
model.add(Dense(3, activation='softmax'))
opt = SGD(learning_rate=0.01)
  
mc = ModelCheckpoint('/content/drive/My Drive/SaturdaysAI/best_model.h5', monitor='accuracy', mode='min', save_best_only=True)
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
# fit network
history=model.fit(X_train, y_train, epochs=epochs,validation_data=(X_test, y_test),batch_size=batch_size, verbose=verbose,callbacks=[mc])
```
